```python
# Criando modelos de Treino
X = dadost.drop(['PassengerId', 'Survived'], axis = 1)
y = dadost.Survived
```

```python
# Fazendo fit X,y
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
# Utilizando KNN
neigh = KNeighborsClassifier(n_neighbors = 3)
```

```python
# Treino com KNN
neigh = neigh.fit(X_train, y_train)
```

```python
# Previsão KNN
pred_KNN = neigh.predict(X_val)
```

```python
# Utilizando Regressão Logistica
clflog = LogisticRegression(random_state = 0, max_iter = 100)
```

```python
# Treino Regressão Logistica
clflog = clflog.fit(X_train, y_train)
```

```python
# Previsão Regressão Logistica
pred_clflog = clflog.predict(X_val)
```

```python
# Utilizando Árvore de Regressão
clfArvore = tree.DecisionTreeClassifier(random_state = 0)
```

```python
# Treino Árvore de Regressão
clfArvore = clfArvore.fit(X_train, y_train)
```

```python
# Previsão Árvore de Regressão
pred_clfArvore = clfArvore.predict(X_val)
```

```python
# Accuracy KNN
round(accuracy_score(y_val, pred_KNN), 2) * 100
```

```python
# Accuracy Regressão Logística
round(accuracy_score(y_val, pred_clflog), 2) * 100
```

```python
# Accuracy Árvore de Regressão
round(accuracy_score(y_val, pred_clfArvore), 2) * 100
```
